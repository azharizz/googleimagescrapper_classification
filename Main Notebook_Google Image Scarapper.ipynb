# **Importing Library and Directory**

```python
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import selenium
from PIL import Image
from urllib import request
from io import BytesIO
```

```python
import splitfolders

# Split with a ratio.
# To only split into training and validation set, set a tuple to `ratio`, i.e, `(.8, .2)`.
splitfolders.ratio(os.getcwd()+'/photos/', output=os.getcwd()+"\\tmp\\", seed=1337, ratio=(.8, .2)) # default 
```

```python
base_dir = os.getcwd()+'/tmp/'
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'val')
```

```python
train_cat_dir = os.path.join(train_dir, 'british_cat')
train_bird_dir = os.path.join(train_dir, 'love_bird')
train_fish_dir = os.path.join(train_dir, 'koi_fish')
validation_cat_dir = os.path.join(validation_dir, 'british_cat')
validation_bird_dir = os.path.join(validation_dir, 'love_bird')
validation_fish_dir = os.path.join(validation_dir, 'koi_fish')
```

# **Modelling**

```python
train_datagen = ImageDataGenerator(
                    rescale=1./255,
                    rotation_range=20,
                    horizontal_flip=True,
                    shear_range = 0.2,
                    fill_mode = 'nearest')

test_datagen = ImageDataGenerator(
                    rescale=1./255,
                    validation_split=0.2)
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir,  
        target_size=(150, 150),  
        batch_size=32,
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        validation_dir, 
        target_size=(150, 150), 
        batch_size=32, 
        class_mode='categorical')
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(3, activation = 'softmax')
])
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=tf.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.fit(
      train_generator,
      # steps_per_epoch=25,
      epochs=20,
      batch_size=32,
      validation_data=validation_generator, 
      # validation_steps=5,
      verbose=2)
```

# **Image Test From Different Web (Not Google)** 

```python
from PIL import Image
from urllib import request
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.preprocessing import image
import numpy as np


url = "https://images.tokopedia.net/img/cache/900/VqbcmM/2020/11/16/be3233a9-5c10-4573-b9ce-bf048166c478.jpg"
res = request.urlopen(url).read()
Sample_Image = Image.open(BytesIO(res)).resize((150,150))

plt.imshow(Sample_Image)
x = image.img_to_array(Sample_Image)
x = np.expand_dims(x, axis=0)

images = np.vstack([x])
classes = model.predict(images, batch_size=10)
# model.summary()
print(classes[0:10])
  

if classes[0][0]==1:
    print('British Cat')
elif classes[0][1]==1:
    print('Love Bird')
elif classes[0][2]==1:
    print('Koi Fish')
else:
    print('error')
```
